```python
import moviepy.editor as mp
import os

def convert_gif_to_mp4(input_file):
    if input_file.lower().endswith('.gif'):  # Only process GIF files
        clip = mp.VideoFileClip(input_file)
        output_file = os.path.splitext(input_file)[0] + '.mp4'  # Set the output file name with .mp4 extension
        clip.write_videofile(output_file, codec='libx264')  # Use the libx264 codec for MP4 encoding
        print(f"Converted {input_file} to {output_file}")
    else:
        print(f"Skipping non-GIF file: {input_file}")

# Specify your file(s) here
files_to_convert = ['/content/animation.gif']

# Convert each file
for file in files_to_convert:
    convert_gif_to_mp4(file)
```
